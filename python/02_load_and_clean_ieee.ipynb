```python
import pandas as pd
import numpy as np

print("Libraries loaded!")
```

```python
transaction = pd.read_csv('C:/Users/aryan/OneDrive/Desktop/Digital Payment Fraud Analysis/IEEE-Fraud/train_transaction.csv')
identity = pd.read_csv('C:/Users/aryan/OneDrive/Desktop/Digital Payment Fraud Analysis/IEEE-Fraud/train_identity.csv')

print(f"Transaction shape: {transaction.shape}")
print(f"Identity shape: {identity.shape}")
```

```python
null_percent = transaction.isnull().sum() / len(transaction) * 100
print("Columns with most missing data:")
print(null_percent.sort_values(ascending=False).head(20))
```

```python
null_percent = transaction.isnull().sum() / len(transaction) * 100
print(f"Total columns: {len(transaction.columns)}")
print(f"Columns with over 50% nulls: {(null_percent > 50).sum()}")
print(f"Columns with over 75% nulls: {(null_percent > 75).sum()}")
print(f"Columns with over 90% nulls: {(null_percent > 90).sum()}")
print(f"Columns with zero nulls: {(null_percent == 0).sum()}")
```

```python
# Drop columns with more than 50% missing values
threshold = 50
cols_to_drop = null_percent[null_percent > threshold].index
transaction_cleaned = transaction.drop(columns=cols_to_drop)

print(f"Columns before cleaning: {transaction.shape[1]}")
print(f"Columns dropped: {len(cols_to_drop)}")
print(f"Columns remaining: {transaction_cleaned.shape[1]}")
```

```python
null_percent_id = identity.isnull().sum() / len(identity) * 100
print(f"Identity table columns: {len(identity.columns)}")
print(f"Columns with over 50% nulls: {(null_percent_id > 50).sum()}")
print(f"Columns with zero nulls: {(null_percent_id == 0).sum()}")

# Drop high null columns from identity too
cols_to_drop_id = null_percent_id[null_percent_id > 50].index
identity_cleaned = identity.drop(columns=cols_to_drop_id)

print(f"Identity columns remaining: {identity_cleaned.shape[1]}")
```

```python
# Merge transaction and identity on TransactionID
df_ieee = transaction_cleaned.merge(identity_cleaned, on='TransactionID', how='left')

print(f"Transaction rows: {len(transaction_cleaned):,}")
print(f"Merged dataset shape: {df_ieee.shape}")
print(f"Fraud cases: {df_ieee['isFraud'].sum():,}")
print(f"Fraud rate: {df_ieee['isFraud'].mean()*100:.4f}%")
```

```python
df_ieee.to_csv('C:/Users/aryan/OneDrive/Desktop/Digital Payment Fraud Analysis/data/processed/ieee_cleaned.csv', index=False)
print(f"IEEE cleaned data saved!")
print(f"Final shape: {df_ieee.shape}")
```
